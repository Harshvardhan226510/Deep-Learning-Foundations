```python
import numpy as np

docs = ['go india',
        'india india',
        'hip hip hurray',
        'jeetega bhai jeetega india jeetega',
        'bharat mata ki jai',
        'kohli kohli',
        'sachin sachin',
        'dhoni dhoni',
        'modi ji ki jai',
        'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer() # oov_token means if any new input is passed while prediction but is not present in the data on which it is trained so the new word will be replaced by oov_token
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
```

```python
sequences
```

```python
len(tokenizer.word_counts)
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
model = Sequential()

model.add(Embedding(18, output_dim=2, input_shape=(5,)))
model.summary()
```

```python
model.compile('adam', 'accuracy')
```

```python
pred = model.predict(sequences)
pred
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data()
```

```python
x_train
```

```python
x_train.shape
```

```python
len(x_train[0])
```

```python
x_train = pad_sequences(x_train, padding='post', maxlen=70)
x_test = pad_sequences(x_test, padding='post', maxlen=70)
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Embedding(88587, 2, input_shape=(50,)))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5)
```
