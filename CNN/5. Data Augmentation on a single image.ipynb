```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/happy-samoyed-dog-field-flowers.jpg', target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='NewImages'):

  i += 1

  if i > 10:
    break
```

```python
img2 = image.load_img('/content/NewImages/_0_633.png', target_size=(200,200))
plt.imshow(img2)
```
