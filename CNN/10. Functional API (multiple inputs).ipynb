```python
from keras.layers import *
from keras.models import Model
from keras.utils import plot_model
```

```python
input_A = Input(shape = (32,))
input_B = Input(shape = (128,))

hidden_1_1 = Dense(8, activation = 'relu')(input_A)
hidden_1_2 = Dense(4, activation = 'relu')(hidden_1_1)

hidden_2_1 = Dense(64, activation = 'relu')(input_B)
hidden_2_2 = Dense(32, activation = 'relu')(hidden_2_1)
hidden_2_3 = Dense(4, activation = 'relu')(hidden_2_2)

concat = concatenate([hidden_1_2, hidden_2_3])

dense = Dense(2, activation = 'relu')(concat)
output = Dense(1, activation = 'linear')(dense)

model = Model(inputs = [input_A, input_B], outputs=output)
```

```python
plot_model(
    model,
    dpi=60,
)
```
