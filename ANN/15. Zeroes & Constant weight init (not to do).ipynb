```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons
```

```python
x, y = make_circles(n_samples=1000, noise=0.1, factor=0.2, random_state=42)
```

```python
plt.scatter(x = x[:, 0], y = x[:, 1], c=y)
```

# **Initializing weights and biases as zero**

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense
from keras import Sequential
```

```python
model = Sequential()

model.add(Dense(2, activation='sigmoid', input_dim=2))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0] = np.zeros(model.get_weights()[0].shape)
initial_weights[1] = np.zeros(model.get_weights()[1].shape)
initial_weights[2] = np.zeros(model.get_weights()[2].shape)
initial_weights[3] = np.zeros(model.get_weights()[3].shape)
```

```python
model.set_weights(initial_weights)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x, y, epochs=100, validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x, y, clf=model)
```

# **Initializing weights & biases a non-zero constant**

```python
x, y = make_moons(n_samples=1000, random_state=2, noise=0.2)
```

```python
plt.scatter(x = x[:, 0], y = x[:, 1], c=y)
```

```python
model = Sequential()

model.add(Dense(2, activation='relu', input_dim=2))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0] = np.ones(model.get_weights()[0].shape)*0.4
initial_weights[1] = np.ones(model.get_weights()[1].shape)*0.4
initial_weights[2] = np.ones(model.get_weights()[2].shape)*0.4
initial_weights[3] = np.ones(model.get_weights()[3].shape)*0.4
```

```python
model.set_weights(initial_weights)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x, y, epochs=100, validation_split=0.2)
```

```python
plot_decision_regions(x, y, clf=model)
```
