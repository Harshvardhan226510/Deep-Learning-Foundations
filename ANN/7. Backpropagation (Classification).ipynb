```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,1], [7,9,1], [6,10,0], [5,5,0]], columns=['cgpa', 'profile_score','placed'])
```

```python
# In thos function you provide your architecture of your neural network to the function
# In our case the the architecture is [2, 2, 1] i.e. 2 input nodes, 2 hidden layer nodes, 1 output node
# The function will calculate the number of trainable parameters in our case there are 9 trainable parameters
# 6 weights and 3 biases
# The function will also initialize the values of the weights and the biases

def initialize_parameters(layer_dims):
  np.random.seed(3)
  parameters = {}
  L  = len(layer_dims)
  for i in range(1, L):

    parameters['W' + str(i)] = np.ones((layer_dims[i-1], layer_dims[i])) * 0.1
    parameters['b' + str(i)] = np.zeros((layer_dims[i], 1))

  return parameters
```

```python
def sigmoid(z):
  A = 1/ (1 + np.exp(-z))
  return A
```

```python
def linear_forward(A_prev, W, b):
  Z = np.dot(W.T, A_prev) + b
  A = sigmoid(Z)
  return A
```

```python
# Forward propogation

def L_layer_forward(X, parameters):

  A = X
  L = len(parameters) // 2

  for l in range (1, L+1):
    A_prev = A
    Wl = parameters['W' + str(l)]
    bl = parameters['b' + str(l)]

    # print("A" + str(l)+": ", A_prev)
    # print("W" + str(l)+": ", Wl)
    # print("b" + str(l)+": ", bl)
    # print("*"*20)

    A = linear_forward(A_prev, Wl, bl)
    # print("A" + str(l)+": ", A)
    # print("-"*20)

  return A, A_prev
```

```python
def update_parameters(parameters, y, y_hat, A1, X):

  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * (y - y_hat) * A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * (y - y_hat) * A1[1][0])
  parameters['b2'][0][0] = parameters['b2'][0][0] + (0.001 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0] * (1 - A1[0][0])* X[0][0])
  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0] * (1 - A1[0][0])* X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0] * (1 - A1[0][0]))

  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * (y - y_hat) * parameters['W2'][1][0] * A1[1][0] * (1 - A1[1][0])* X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * (y - y_hat) * parameters['W2'][1][0] * A1[1][0] * (1 - A1[1][0])* X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * (y - y_hat) * parameters['W2'][1][0] * A1[1][0] * (1 - A1[1][0]))
```

```python
parameters = initialize_parameters([2,2,1])
epochs = 100

for i in range(epochs):

  Loss = []

  for j in range(df.shape[0]):

    x = df[['cgpa', 'profile_score']].values[j].reshape(2,1)
    y = df[['placed']].values[j][0]

    # Initializing parameters

    y_hat, A1 = L_layer_forward(x , parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters, y, y_hat, A1, x)

    Loss.append(-y*np.log(y_hat) - (1-y)*np.log(1-y_hat))

  print('Epoch :==: ', i+1, 'Loss :==: ', np.array(Loss).mean())

parameters
```
